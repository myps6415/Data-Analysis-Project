```python
import requests
import datetime
import pandas as pd

date = datetime.datetime.now()
datestr = date.strftime("%Y%m%d")

res = requests.get('https://www.twse.com.tw/exchangeReport/MI_5MINS_INDEX?response=json&date={}&_=1570003179176'.format(datestr))
```

```python
df = pd.DataFrame(res.json()['data'], columns=res.json()['fields'])
df.head()
```

```python
df = df.set_index('時間')
df.head()
```

```python
df.index = pd.to_datetime(datestr + ' ' + df.index)
df.head()
```

```python
df = df.applymap(lambda s: float(str(s).replace(',', '')))
df.head()
```
