```python
import requests
import pandas as pd
from io import StringIO

# 月報網址
url = 'https://mops.twse.com.tw/nas/t21/sii/t21sc03_110_1_0.html'
r = requests.get(url)
r.encoding = 'big5'
```

```python
dfs = pd.read_html(r.text)
print(len(dfs))
```

```python
not_one_list = []

for i in range(0, len(dfs)):
    if len(dfs[i]) != 1:
        not_one_list.append(i)

print(not_one_list)
```

```python
df = pd.concat([dfs[i] for i in not_one_list])
print(len(df))
df.head()
```

```python
df.columns = [column[1] for column in df.columns]
print(len(df))
df.head()
```

```python
df = df[df['備註'].notnull()]
print(len(df))
df.head()
```
