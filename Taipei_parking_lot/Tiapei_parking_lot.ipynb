```python
import pandas as pd
import json

with open('/content/drive/MyDrive/Colab Notebooks/TCMSV_alldesc.txt') as f:
    data = json.load(f)

print(type(data))
data
```

```python
df = pd.DataFrame(data['data']['park'])
print(len(df))
df.head()
```

```python
df = df[(df['area'] == '信義區') | (df['area'] == '大安區') | (df['area'] == '松山區')]
df = df.reset_index().drop(columns=['index'])
print(len(df))
df.head()
```

```python
for enu, i in enumerate(df['payex']):
  print(i)
  if enu == 10:
    break
```

```python
df['payex'][28]
```

```python
import re
payex = []

for enu, i in enumerate(df['payex']):
  if re.search(r'(月租|月票).*', i):
    payex.append(re.search(r'(月租|月票).*', i).group())
  else:
    payex.append(None)

df['payex'] = payex
df.head()
```

```python
df = df[~df['payex'].isnull()]
print(len(df))
df.head()
```

```python
for enu, payex_ in enumerate(df['payex']):
  print(enu, re.search(r'\d+', re.search(r'\w.+', payex_.replace(',','')).group()).group())
```
